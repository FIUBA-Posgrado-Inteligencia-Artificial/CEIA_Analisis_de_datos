### Discretización

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import KBinsDiscretizer
```

```python
# Cargamos el dataset de ejemplo (penguins)
penguins = sns.load_dataset("penguins").dropna()

# split 80%/20% (sin target definido)
df, test_df = train_test_split(penguins, test_size=0.2, random_state=42)

print(df.shape, test_df.shape)
```

```python
df.head()
```

```python
# Seleccionamos una variable continua para discretizar
variable = "bill_length_mm"
```

```python
# Visualizamos la distribución original
fig, ax = plt.subplots(figsize=(8,4))

sns.histplot(df[variable], bins=20, kde=True, ax=ax)
ax.set_title("Distribución Original")
plt.show()
```

#### Función auxiliar para mostrar resultados

```python
import matplotlib.patches as mpatches
def plot(new_var, boxplot=False, div=False):
    fig, axes = plt.subplots(1, 3, figsize=(14, 3))
    
    # Histograma
    sns.histplot(data=df, x=variable, hue=new_var, multiple="stack",
                 bins=28, palette="Set2", ax=axes[0], legend=False)
    axes[0].set_title(f"Histograma por categoría ({new_var})")
    axes[0].set_xlabel(variable)
    axes[0].set_ylabel("Frecuencia")
    axes[0].grid(axis='y', ls='--')

    # Boxplot
    if(boxplot):
        sns.boxplot(data=df, x=variable,
           y=[""] * len(df), fliersize=0,
           color="#e0e0e0", ax=axes[1],
           boxprops=dict(alpha=0.5), width=0.4)
    
    # Swarmplot
    sns.swarmplot( data=df, x=variable,
        y=[""] * len(df),  # fuerza a que todos compartan la misma fila
        hue=new_var, palette="Set2", ax=axes[1], legend=False, size=4)
    axes[1].set_title("Discretización")
    axes[1].set_xlabel(variable)
    axes[1].set_ylabel("")

    # Countplot
    sns.countplot(x=new_var, data=df, hue=new_var, palette="Set2", 
                  ax=axes[2], legend=False)
    axes[2].set_xlabel("Categoría")
    axes[2].set_ylabel("Cantidad de pingüinos")
    axes[2].set_title("Distribución de pingüinos según longitud de pico")
    axes[2].grid(axis='y', ls='--')

    # líneas verticales de separación de grupos
    cortes = df.groupby(new_var, observed=False)[variable].max().sort_values().values
    for corte in cortes[:-1]:
        if(div): # div en histograma
            axes[0].axvline(x=corte, color='magenta', linestyle='--', alpha=0.7, linewidth=1.5)
        axes[1].axvline(x=corte, color='magenta', linestyle='--', alpha=0.7, linewidth=1.5)
    
    # leyenda única (compartida)
    categorias = sorted(df[new_var].dropna().unique())
    colores = sns.color_palette("Set2", len(categorias))
    handles_manuales = [
        mpatches.Patch(color=colores[i], label=str(cat)) 
        for i, cat in enumerate(categorias)]
    
    fig.legend(handles=handles_manuales, 
        title=new_var, loc='center left',
        bbox_to_anchor=(1.01, 0.5))

    plt.tight_layout()
    plt.show()
```

#### Por intervalos iguales

```python
# Discretización por intervalos de igual amplitud
df["equal_width"] = pd.cut(df[variable], bins=4, labels=["Corto", "Medio", "Largo", "Muy Largo"])
df.head(10)
```

```python
plot('equal_width', div=True)
```

#### Por intervalos de igual frecuencia

Pensado para EDA.

```python
# Discretización por intervalos de igual frecuencia
df["equal_freq"], bins = pd.qcut(df[variable], q=4, labels=["Muy Corto", "Corto", "Medio", "Largo"], retbins=True)
df.head(10)
```

```python
plot('equal_freq', True)
```

#### Discretización por puntos de corte

Pensado para ML (ya devuelve los códigos, no hay que codificar).

```python
# Cuantiles
kbin_quantile = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='quantile')
df["kbin_quantile"] = kbin_quantile.fit_transform(df[[variable]]).astype(int)
df.head(10)
```

```python
plot('kbin_quantile', True)
```

#### Con K-means

```python
# K-means
kbin_kmeans = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='kmeans')
df["kbin_kmeans"] = kbin_kmeans.fit_transform(df[[variable]]).astype(int)
df.head(10)
```

```python
plot('kbin_kmeans', True)
```

#### Rango fijo

```python
# Forzamos los cortes manuales, mayores y menores de 50 mm
cortes = [-np.inf, 50, np.inf]
etiquetas = ["<= 50 mm", "> 50 mm"]

df["grupo_50"] = pd.cut(df[variable], bins=cortes, labels=etiquetas, include_lowest=True)
```

```python
plot('grupo_50')
```

#### Comparación de resultados

```python
# Resultados
penguins_disc = df[[variable, "equal_width", "kbin_quantile", "kbin_kmeans", "grupo_50"]]
penguins_disc
```

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

cols = ["equal_width", "kbin_quantile", "kbin_kmeans", "grupo_50"]
titles = ["Intervalos iguales", "Puntos de corte", "K-means", "Ancho fijo"]

for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i])
    axes[i].set_title(titles[i])
    axes[i].set_xlabel("Categoría")
    axes[i].set_ylabel("Cantidad")

fig.tight_layout()
plt.show()
```
