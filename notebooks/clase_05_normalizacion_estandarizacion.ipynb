### Normalización y estandarización

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# Cargo el dataset de penguins
df = sns.load_dataset('penguins').dropna(subset=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g'])

# Selecciono algunas columnas numéricas para la normalización y estandarización
columns = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
df_subset = df[columns]
df_subset.head()
```

```python
df_subset.describe()
```

#### Normalización (Min-Max)

```python
# Normalización (Min-Max)
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df_subset), columns=df_subset.columns, index=df_subset.index)

print("Normalizado:\n", df_normalized.head())
```

```python
df_normalized.describe()
```

#### Estandarización (Z-Score)

```python
# Estandarización (Z-Score)
scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(df_subset), columns=df_subset.columns, index=df_subset.index)
print("\nEstandarizado:\n", df_standardized.head())
```

```python
df_standardized.describe()
```

#### Armamos una muestra de datos para ver cómo quedan normalizados y estandarizados.

```python
# mostramos la columna de ancho del pico
col_muestra = 'bill_depth_mm'

#  incluimos el mínimo y máximo del dataset original
idx_min = df_subset[col_muestra].idxmin()
idx_max = df_subset[col_muestra].idxmax()
idx_random = df_subset.drop([idx_min, idx_max]).sample(7, random_state=42).index
idx_muestra = [idx_min, idx_max] + list(idx_random)

tabla_comparativa = pd.DataFrame({
    'Original': df_subset.loc[idx_muestra, col_muestra],
    'Normalizado': df_normalized.loc[idx_muestra, col_muestra],
    'Estandarizado': df_standardized.loc[idx_muestra, col_muestra]
}).sort_values(by='Original')

tabla_comparativa
```

```python
# Gráfico
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(10, 10))
titles = ["Original", "Normalizado (MinMax)", "Estandarizado (Z-Score)"]

# Recorrer cada variable y graficarla en las tres columnas
for i, col in enumerate(columns):
    sns.histplot(df_subset[col], kde=True, ax=axes[i, 0], color="gray")
    sns.histplot(df_normalized[col], kde=True, ax=axes[i, 1], color="lightgreen")
    sns.histplot(df_standardized[col], kde=True, ax=axes[i, 2], color="red")

    # Etiquetas
    axes[i, 0].set_ylabel(col)
    for j in range(3):
        axes[i, j].set_xlabel("")
        axes[i, j].set_title(titles[j])

# Ajustar el layout
plt.tight_layout()
plt.show()
```

#### Vamos a ver qué pasa con las estadísticas de las variables al normalizar y estandarizar

```python
df_subset.describe().loc[['mean', 'std','min','max']]
```

```python
df_normalized.describe().loc[['mean', 'std','min','max']]
```

```python
df_standardized.describe().loc[['mean', 'std','min','max']].round(6)
```

```python
# calculamos las matrices de covarianza y de correlación (Pearson)
cov_original = df_subset.cov()
cov_normalized = df_normalized.cov()
cov_standardized = df_standardized.cov()

corr_original = df_subset.corr(method='pearson')
corr_normalized = df_normalized.corr(method='pearson')
corr_standardized = df_standardized.corr(method='pearson')
```

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 14))
titles = ["Original", "Normalizado (MinMax)", "Estandarizado (Z-Score)"]

# Covarianza original
sns.heatmap(
    cov_original,
    annot=True,
    cmap="coolwarm",
    fmt=".3f",
    ax=axes[0, 0]
)
axes[0, 0].set_title("Matriz de Covarianza - Datos originales")

# Correlación original
sns.heatmap(
    corr_original,
    annot=True,
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    fmt=".3f",
    ax=axes[0, 1]
)
axes[0, 1].set_title("Correlación de Pearson - Datos Originales")

# Covarianza datos normalizados
sns.heatmap(
    cov_normalized,
    annot=True,
    cmap="coolwarm",
    fmt=".3f",
    ax=axes[1, 0]
)
axes[1, 0].set_title("Matriz de Covarianza - Normalizado (MinMax)")

# Correlación datos normalizados
sns.heatmap(
    corr_normalized,
    annot=True,
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    fmt=".3f",
    ax=axes[1, 1]
)
axes[1, 1].set_title("Correlación de Pearson - Normalizado (MinMax)")

# Covarianza datos estandarizados
sns.heatmap(
    cov_standardized,
    annot=True,
    cmap="coolwarm",
    fmt=".3f",
    ax=axes[2, 0]
)
axes[2, 0].set_title("Matriz de Covarianza - Estandarizado (Z-Score)")

# Correlación datos estandarizados
sns.heatmap(
    corr_standardized,
    annot=True,
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    fmt=".3f",
    ax=axes[2, 1]
)
axes[2, 1].set_title("Correlación de Pearson - Estandarizado (Z-Score)")

plt.tight_layout()
plt.show()
```

#### ¿Qué pasa con las estadísticas de las variables al normalizar y estandarizar?

La media, la varianza y la covarianza de las variables cambian.

La correlación de Pearson permanece igual porque es invariante frente a transformaciones lineales de escala y centrado.
